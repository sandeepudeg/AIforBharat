# Understanding :

1.  An AI agent can basiscaaly be decomposed in to three components :
    - Reasoning model : Which is basically a LM
    - Actionable tools : Tools that the agent can use to perform tasks.
    - The governing Orchestration Layer : This layer decides which tool to call and how to orchestrate them.

2.  There are two types of agents:
    - Reactive Agent : These agents do not have any memory or history, they only take into account the current state of the environment.
    - Proactive Agent : These agents have some form of memory or history, they take into account past states as well as the current state.

```python
from google.adk.agents import Agent
from google.adk.runners import InMemoryRunner
from google.adk.tools import google_search
from google.genai import types

print("✅ ADK components imported successfully.")
```

```python
root_agent = Agent(
    name="helpful_assistant",
    model="gemini-2.5-flash-lite",
    description="A simple agent that can answer general questions.",
    instruction="You are a helpful assistant. Use Google Search for current info or if unsure.",
    tools=[google_search],
)

print("✅ Root Agent defined.")
```

```python
runner = InMemoryRunner(agent=root_agent)

print("✅ Runner created.")
```

```python
response = await runner.run_debug("What is Agent Development Kit from Google? What languages is the SDK available in?")
print("✅ Response received:", response)
```

```python
response = await runner.run_debug("what is teh traffic cobnditions in hinjewadi Pune")
print("✅ Response received:", response)
```
