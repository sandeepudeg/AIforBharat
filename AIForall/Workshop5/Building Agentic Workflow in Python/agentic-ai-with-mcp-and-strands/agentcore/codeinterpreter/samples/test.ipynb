```python
import os
from dotenv import load_dotenv

# Load variables from .env into the system environment
load_dotenv()

# (Optional) Verify they loaded as strings
print(f"Region: {os.environ.get('AWS_DEFAULT_REGION')} (Type: {type(os.environ.get('AWS_DEFAULT_REGION')).__name__})")
```

```python
print(f"NASA: {os.environ.get('NASA_API_KEY')}) (Type: {type(os.environ.get('NASA_API_KEY')).__name__})")
```

```python
import boto3

def check_identity():
    try:
        sts = boto3.client('sts')
        identity = sts.get_caller_identity()
        print(f"✅ Authenticated as: {identity['Arn']}")
    except Exception as e:
        print(f"❌ Authentication failed: {e}")

check_identity()
```
